```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from textblob import TextBlob
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
```

```python
text = "The movie was amazing and I really liked it"

analysis = TextBlob(text)

print("Polarity:", analysis.sentiment.polarity)
print("Subjectivity:", analysis.sentiment.subjectivity)
```

```python
import nltk
text = """
Natural Language Processing is a field of artificial intelligence that helps computers
understand human language. It is widely used in applications like chatbots, translation,
sentiment analysis and text summarization.
"""
nltk.download('punkt_tab')
parser = PlaintextParser.from_string(text, Tokenizer("english"))
summarizer = LsaSummarizer()

summary = summarizer(parser.document, 2)

for sentence in summary:
    print(sentence)
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

text = """
Natural Language Processing is a field of artificial intelligence that helps computers
understand human language. It is widely used in applications like chatbots, translation,
sentiment analysis and text summarization.
"""

parser = PlaintextParser.from_string(text, Tokenizer("english"))
summarizer = LsaSummarizer()

summary = summarizer(parser.document, 2)

for sentence in summary:
    print(sentence)
```
