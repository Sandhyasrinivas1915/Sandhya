```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
text="Hi i am emily.Nice to meet u"
output1=word_tokenize(text)
output2=sent_tokenize(text)
print(output1)
print(output2)
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
input="this is the second session"
doc=nlp(input)
```

```python
for token in doc:
  print(token)
```

```python
for sent in doc.sents:
  print(sent)
```
