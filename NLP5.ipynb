```python
import pandas as pd
import numpy as np
import re
import nltk
nltk.download('punkt_tab')
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
parah = "Space is a three-dimensional continuum containing positions and directions.[1] In classical physics, physical space is often conceived in three linear dimensions. Modern physicists usually consider it, with time, to be part of a boundless four-dimensional continuum known as spacetime.[2] The concept of space is considered to be of fundamental importance to an understanding of the physical universe. However, disagreement continues between philosophers over whether it is itself an entity, a relationship between entities, or part of a conceptual framework."
```

```python
sentence = nltk.sent_tokenize(parah)
```

```python
sentence
```

```python
x = WordNetLemmatizer()
```

```python

for i in range(len(sentence)) :
  words = nltk.word_tokenize(sentence[i])
  nwords = [x.lemmatize(word) for word in words]
  sentence[i] = ' '.join(nwords)
```

```python
sentence
```

```python
sentence
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
for i in range(len(sentence)):
  words = nltk.word_tokenize(sentence[i])
  words = [stemmer.stem(word) for word in words]
  sentence[i] = ' '.join(words)
```
